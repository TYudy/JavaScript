# Información básica de dataset

Link de descarga dataset: https://www.kaggle.com/datasets/shashwatwork/depression-and-mental-health-data-analysis

Con respecto al conjunto de datos RHMCD-20, incluye información de una amplia gama de fuentes, que incluyen adolescentes de Bangladesh, estudiantes universitarios, amas de casa, profesionales de empresas y corporaciones, y otras personas. Estos datos de encuesta se recopilaron para el análisis de datos sobre depresión y salud mental.

Preguntas de la encuesta:

- Edad(Age): Representa la edad de los participantes.

- Género(Gender): Indica el género de los participantes.

- Ocupación(Occupation): Representa las ocupaciones de los participantes.

- Días en Interiores(Days_Indoors): Indica la cantidad de días que el participante no ha salido de casa.

- Aumento de Estrés(Growing_Stress): Indica si el estrés del participante está aumentando día a día (Sí/No).

- Frustración en Cuarentena(Quarantine_Frustration): Frustraciones en las
primeras dos semanas de cuarentena (Sí/Tal vez/No).

- Cambios en Hábitos(Changes_Habits): Representa cambios importantes en los hábitos alimenticios y de sueño (Sí/Tal vez/No).

- Historial de Salud Mental(Mental_Health_History): Antecedentes de trastornos mentales en la generación anterior (Sí/No).

- Cambio de Peso(Weight_Change): Destaca cambios en el peso corporal durante la cuarentena (Sí/Tal vez/No).

- Altibajos del Estado de Ánimo(Mood_Swings): Representa cambios extremos de humor (Bajo/Medio/Alto).

- Dificultades para Afrontar(Coping_Struggles): La incapacidad para afrontar problemas diarios o estrés (Sí/Tal vez/No).

- Interés en el Trabajo(Work_Interest): Representa si el participante está perdiendo interés en trabajar (Sí/No).

- Debilidad Social(Social_Weakness): Comunica sentirse mentalmente débil al interactuar con los demás (Sí/No).
